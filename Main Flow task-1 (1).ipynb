# Data Analysis Question :

```python
# 1) what is the average score in math(G3)?
avg_g3=df['G3'].mean()
print("The average score in math (G3):",avg_g3)
```

```python
# 2) how many students scored above 15 in their final score(G3)?
above_15=(df["G3"]>15).sum()
print("The total amount of student who scored above 15:",above_15)
```

```python
# 3) Is there a correlation between study time(study time) and the final grade(G3)?
correlation=df["studytime"].corr(df["G3"])
print("The correlation between studytime and final grade:",correlation)
```

```python
# 4) which gender has a higher average final grade(G3)?
avg_gender=df.groupby("sex")["G3"].mean()
higher_gender=avg_gender.idxmax()
print("gender with higher average final grade:",higher_gender)
```

# Data Visualization :

# Histogram

Plot a histogram of final grades (G3):
  1. Make a histogram to show how the final grades (G3) are spread out.  
2. It helps to see if most students got low, medium, or high marks



```python
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=15, kde=True, color='skyblue')
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

# Scatter plot :

Create a scatter plot between study time and final grade (G3):
  1. Make a scatter plot to show the connection between study time and final grade (G3).
     
  2. It helps to see if studying more leads to higher marks.


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='studytime', y='G3', hue='sex', palette='Set2')
plt.title('Study Time vs Final Grade (G3)')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.tight_layout()
plt.show()
```

# Barplot :

Comparison of Average Final Grades by Gender in barplot

  1.Make a bar chart to compare the average final grades of male and female students.
  
  2.This helps to see which group scored higher on average.

```python
plt.figure(figsize=(8, 5))
avg_scores = df.groupby('sex')['G3'].mean().reset_index()
sns.barplot(data=avg_scores, x='sex', y='G3', palette='pastel')
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.tight_layout()
plt.show()
```
